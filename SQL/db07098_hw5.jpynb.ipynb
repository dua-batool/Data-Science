```python
import seaborn as sns                      
import matplotlib.pyplot as plt             
%matplotlib inline  
```

```python
a = df['jobrole'].value_counts()
print(a)
ax = df['jobrole'].value_counts().plot(kind='bar', figsize=(10,4))
ax.set_xlabel("Job Roles")
ax.set_ylabel("Counts")
plt.show()
```

According to the bar graph, the most frequently occurring job role is "Sales-Executive," noticeably towering over 300, implying its prevalence within the dataset. Following closely are the roles of "Research Scientist" and "Laboratory Technician," exhibiting bars ranging from 250 to 300, indicating their substantial representation as well. In contrast, the remaining job roles collectively demonstrate bars indicating counts below 150, suggesting their relatively lower occurrence. Notably, the dataset comprises a total of 9 distinct job roles, each contributing to the overall employment landscape depicted in the graph.

```python
ax = df['department'].hist()
ax.set_xlabel("Job Roles")
ax.set_ylabel("Counts") 
plt.show()
```

Based on the histogram data, it is evident that the company operates across three primary departments: Sales, Research & Development, and Human Resources. The "Research & Development" department notably accommodates the highest number of employees, reflecting its significant workforce contribution within the company. Conversely, the Human Resources department appears to have a comparably smaller employee count, measuring less than 100, as indicated in the graph. The Research department hosts a workforce figure closer to 1000, but slightly below that mark. On the other hand, the Sales department demonstrates a robust employment presence, surpassing 400 employees in the company.

```python
sns.lmplot(data=df,x="age", y="monthlyincome",line_kws={'color': 'red'},scatter=True)
```

The scatter plot reveals a clear pattern: as employees get older ("Age" increases), their monthly income ("MonthlyIncome") tends to go up as well. This suggests that older employees generally earn more.
