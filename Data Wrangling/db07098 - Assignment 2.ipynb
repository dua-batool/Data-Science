### Droping columns with null values

```python
formatted = formatted.dropna(axis=1, how='all')
```

```python
formatted.head()
```

### Replacing missing values

For categorical data in our dataset, we can replace the missing values with mode because mode represents the most common category. 

```python
formatted['Country of Citizenship'] = formatted['Country of Citizenship'].fillna(formatted['Country of Citizenship'].mode()[0])
formatted['Region'] = formatted['Region'].fillna(formatted['Region'].mode()[0])
formatted['City / Town / District'] = formatted['City / Town / District'].fillna(formatted['City / Town / District'].mode()[0])
formatted['Laptop Type'] = formatted['Laptop Type'].fillna(formatted['Laptop Type'].mode()[0])
formatted['Are you employed full-time while attending Mason?'] = formatted['Are you employed full-time while attending Mason?'].fillna(formatted['Are you employed full-time while attending Mason?'].mode()[0])
formatted['What is your approximate current level of programming skill in Python?'] = formatted['What is your approximate current level of programming skill in Python?'].fillna(formatted['What is your approximate current level of programming skill in Python?'].mode()[0])
formatted['What is your approximate current level of programming skill in R?'] = formatted['What is your approximate current level of programming skill in R?'].fillna(formatted['What is your approximate current level of programming skill in R?'].mode()[0])
formatted['What is your approximate current level of programming skill in SQL?'] = formatted['What is your approximate current level of programming skill in SQL?'].fillna(formatted['What is your approximate current level of programming skill in SQL?'].mode()[0])
formatted['Undergraduate Degree'] = formatted['Undergraduate Degree'].fillna(formatted['Undergraduate Degree'].mode()[0])

formatted['Expected Graduation date from Mason MS program?'] = formatted['Expected Graduation date from Mason MS program?'].fillna(formatted['Expected Graduation date from Mason MS program?'].mode()[0])
```

In the above code, I chose the columns with categorical data and replaced null values with the mode of that column. Even though the graduation date is not categorical, I chose mode for this column as well because it seemed the most reasonable choice.

```python
formatted = formatted.replace('6l',61)
formatted = formatted.replace('7O',70)
```

Fixed some values in the height column as it was giving some errors on running the code below. both 6l and 7O were strings which was why we couldn't be converted into numeric values so changed 6l to 61 and 7O to 70.

```python
formatted['Height (Inches)'] = pd.to_numeric(formatted['Height (Inches)'])
formatted['Height (Inches)'] = formatted['Height (Inches)'].fillna(formatted['Height (Inches)'].mean())
```

Replacing the missing values of height with mean instead of mode as imputing missing values with the mean can provide a reasonable estimate of the "typical" height.

```python
formatted['What is your goal for learning about data analytics?'] = formatted['What is your goal for learning about data analytics?'].fillna('To make data-driven decisions')
```

I filled the missing values of goal for learning data analytics with a common response.

### Fixing some values


```python
formatted['Age (years)'] = formatted['Age (years)'].replace('2 3',23)
formatted['Age (years)'] = formatted['Age (years)'].replace('1993',30)
```

```python
formatted['Age (years)']
```

The age column had 2 errors. Replaced '2 3' with 23, assuming that the person would have left a space in between by mistake. Replaced 1993 with 30, assuming that the person entered his birth year instead. So calculated the age by 2023 - 1993 = 30.

```python
formatted['Laptop Type'] = formatted['Laptop Type'].replace('2','Apple/MacBook')
formatted['Laptop Type'] = formatted['Laptop Type'].replace('0','Microsoft/Windows')
```

```python
formatted['Laptop Type']
```

There were 2 incorrect values in the Laptop Type column. Replaced the value 0 with Microsoft/Windows and the value 2 with Apple/MacBook.

```python
formatted['Country of Citizenship'] = formatted['Country of Citizenship'].replace('USA','United States')
formatted['Country of Citizenship'] = formatted['Country of Citizenship'].replace('US','United States')
```

Replaced USA and US to United States as they are the same thing.

```python
formatted.to_csv('580SurveyCleanup_cleaned.csv',index=False)
```

```python
cleaned_data= pd.read_csv('580SurveyCleanup_cleaned.csv')
```

```python
question_labels = [
    'Section', 'Gender', 'Age (years)', 'Height (Inches)','Country of Citizenship', 'Undergraduate Degree',
    'Expected Graduation date from Mason MS program?', 'Laptop Type', 'Are you employed full-time while attending Mason?',
    'What is your approximate current level of programming skill in Python?', 'What is your approximate current level of programming skill in R?',
    'What is your approximate current level of programming skill in SQL?', 'What is your goal for learning about data analytics?'
]
height_label='Height (Inches)'
age_label='Age (years)'

descriptions = [cleaned_data[question_labels[i]].describe() for i in range(13)]
freq = {}
for i in range(13): freq[question_labels[i]] = cleaned_data[question_labels[i]].value_counts().index[-1]

medians = {}
for i in range(13): medians[question_labels[i]] = 'NaN'

top = {}
for i in range(13): top[question_labels[i]] = 'NaN'


medians[height_label] = cleaned_data[height_label].astype(float).median()
medians[age_label] = cleaned_data[age_label].astype(float).median()

top[height_label] = cleaned_data[height_label].astype(float).max()
top[age_label] = cleaned_data[age_label].astype(float).max()

descriptions = pd.concat(descriptions, axis=1)
descriptions = descriptions.append(pd.Series(medians, name='Median'))
descriptions = descriptions.append(pd.Series(top, name='Max'))
```
